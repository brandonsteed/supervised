# Predicting Breast Cancer and Identifying Traits Leading to Diagnosis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import time
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv('wi-cancer-data.csv')
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df.corr()
```

```python
df2 = df.drop(['id_number', 'malignant'], axis=1)
```

```python
df2.isnull().sum()
```

```python
df2 = df2.fillna(3.54)
```

```python
df2.isnull().sum()
```

```python
# Normalize the data so that all variables have a mean of 0 and standard deviation
# of 1.
X = StandardScaler().fit_transform(df2)

# The NumPy covariance function assumes that variables are represented by rows,
# not columns, so we transpose X.
Xt = X.T
Cx = np.cov(Xt)
print('Covariance Matrix:\n', Cx)

# Calculating eigenvalues and eigenvectors.
eig_val_cov, eig_vec_cov = np.linalg.eig(Cx)

# Inspecting the eigenvalues and eigenvectors.
for i in range(len(eig_val_cov)):
    eigvec_cov = eig_vec_cov[:, i].reshape(1, 9).T
    print('Eigenvector {}: \n{}'.format(i + 1, eigvec_cov))
    print('Eigenvalue {}: {}'.format(i + 1, eig_val_cov[i]))
    print(40 * '-')

print(
    'The percentage of total variance in the dataset explained by each',
    'component calculated by hand.\n',
    eig_val_cov / sum(eig_val_cov)
)

plt.plot(eig_val_cov)
plt.show()
```

```python
sklearn_pca = PCA(n_components=1)

X_pca = sklearn_pca.fit_transform(X)

print(
    'The percentage of total variance in the dataset explained by each',
    'component from Sklearn PCA.\n',
    sklearn_pca.explained_variance_ratio_
)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=1e9)
y = df['malignant']

# Cross-val mean score and the 95% confidence interval of the score estimate
logit_score = cross_val_score(lr, X_pca, y, cv=10)
print("Accuracy: %0.3f (+/- %0.3f)" % (logit_score.mean(), logit_score.std() * 2))

#Cross-validation
from sklearn.model_selection import cross_val_score
print('\n')
print(cross_val_score(lr, X_pca, y, cv=10))
```

## Logistic Regression Looks Good
Logistic regression provides for a good model fit.  Implementing L1 or L2 regularization did not make any significant difference in the model, which makes sense.  The PCA transformation already addressed multicollineraity from the data.

```python
# Random Forest Classifier
from sklearn import ensemble
rfc = ensemble.RandomForestClassifier()
X = df2

rfc.fit(X, y)

# Cross-val mean score and the 95% confidence interval of the score estimate
rfc_score = cross_val_score(rfc, X, y, cv=10)
print("Accuracy: %0.3f (+/- %0.3f)" % (rfc_score.mean(), rfc_score.std() * 2))

#Cross-validation
from sklearn.model_selection import cross_val_score
print('\n')
print(cross_val_score(rfc, X, y, cv=10))
```

```python
importances = rfc.feature_importances_

std = np.std([tree.feature_importances_ for tree in rfc.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]
# Print the feature ranking
print("Feature ranking:")

for f in range(X.shape[1]):
    print("%d. (%f)  %s" % (f + 1, importances[indices[f]], X.columns[f]))

feature_importance = rfc.feature_importances_
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
plt.subplot(1, 2, 2)
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, X.columns[sorted_idx])
plt.xlabel('Relative Importance')
plt.title('Variable Importance')
plt.show()
```
